```python
import numpy as np
import matplotlib.pyplot as plt
```

 np.linspace() - define a quantidade de pontos de distância igual
 np.arrange() - define o espaçamento entre pontos do limite dado 
 lambda - recebe um argumente e define a função em uma linha
 append - adiciona o item a lista 
 exemplo: k[]

```python
h = 0.1
x = np.arange (-5, 5, h)
f = lambda x: (x**4)+(x**3)+(3*x**2)-25
y = f(x)
d = np.diff(y)/h
x_diff = x[:-1]
f_exato = (4*x_diff**3)+(3*x_diff**2)+(6*x_diff)
```

```python
plt.plot(x,y,c ='purple')
plt.plot(x_diff,d, 'c--')
plt.plot(x_diff, f_exato, '.', c="pink")
plt.show()
```

```python
h = 0.1
x = np.arange (-5, 5, h)
f = lambda x:(7*x**3-5*x+1)/(2*x**4+x**2+1)
y = f(x)
d = np.diff(y)/h
x_diff = x[:-1]
f_exato = lambda x:((-14*x**6+37*x**4-8*x**3+26*x**2-2*x-5)/((2*x**4+x**2+1)**2))
y_exato = f_exato(x)
```

```python
plt.plot(x,y, c = 'purple',
        label = "Função exata")
plt.plot(x_diff,d,'--',c = 'violet',
         label = 'Derivada numérica')
plt.plot(x,y_exato, '.', c = 'pink',
         label = 'Solução exata')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.001
x = np.linspace(-np.pi, np.pi, 100)
f = lambda x: np.cos(x)+np.sin(2*x)
y = f(x)
```

```python
plt.plot (x,y, c ='pink')
plt.show()
```

```python
from scipy.misc import derivative
```

```python
a0 = f(0)
a1 = derivative (f, 0, h, n = 1)
a2 = derivative (f, 0, h, n = 2) / 2
a3 = derivative (f, 0, h, n = 3, order = 5) / 6
a4 = derivative (f, 0, h, n = 4, order = 5) / 24
```

```python
print (a0, a1, a2, a3, a4)
```

```python
f_nova = lambda x: a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4
T4 = f_nova(x)
```

```python
plt.rcParams.update({'font.size': 14})
plt.plot(x,y, c='pink',
        label=r"$F(x)$")
plt.plot(x,T4,'--', c='violet',
        label=r"$T{4}(x)$")
plt.xlim([-np.pi,np.pi])
plt.ylim([-np.pi,np.pi])
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()
```
